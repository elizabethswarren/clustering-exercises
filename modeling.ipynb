```python
# imports
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

from pydataset import data


```

```python
iris = data('iris')
```

```python
iris.head()
```

```python
iris = iris.rename(columns={'Sepal.Length': 'sepal_length', 'Sepal.Width': 'sepal_width', 'Petal.Length': 'petal_length', 'Petal.Width':'petal_width', 'Species':'species'})
```

```python
iris.head()
```

## 1. Clustering with the Iris Dataset

##       a. Choose features other than the ones used in the lesson.

```python
petal_cluster = iris[['petal_length', 'petal_width']]
```

```python
petal_cluster
```

```python
kmeans = KMeans(n_clusters=3)

kmeans.fit(petal_cluster)

kmeans.predict(petal_cluster)
```

```python
iris['petal_cluster'] = kmeans.predict(petal_cluster)
```

```python
iris.head()
```

```python
kmeans.cluster_centers_
```

```python
centroids = pd.DataFrame(kmeans.cluster_centers_, columns = petal_cluster.columns[:2])
centroids
```

## b. Visualize the results of your clustering.

### Visualization of original data

```python
sns.scatterplot(iris, x = 'petal_width', y = 'petal_length', hue='species', palette = 'viridis')

plt.title('Visualizing Actual Species')

plt.show()
```

### Visualization of cluster

```python
sns.scatterplot(iris, x = 'petal_width', y = 'petal_length', hue = 'petal_cluster', palette = 'viridis')

centroids.plot.scatter(x = 'petal_width', y = 'petal_length', ax = plt.gca(), color ='k', alpha = 0.3, s = 100, marker = (7,1,0), label = 'centroids')

plt.title('Visualizing Clusters with Centroids')

plt.show()
```

### c. Use the elbow method to visually select a good value for k.

```python
inertia_table = []

for n in range (1, 21):
    kmeans = KMeans(n_clusters=n)
    kmeans.fit(petal_cluster)
    inertia_table.append(kmeans.inertia_)
    
```

```python
plt.plot(np.arange(1,21,), inertia_table)

plt.xticks(np.arange(1,21,step=1))

plt.title('Change in Inertia')

plt.show()
```

## 2. Use the techniques discussed in this lesson, as well as the insights gained from the exploration exercise to perform clustering on the mall customers dataset. Be sure to visualize your results!

```python
mall = pd.read_csv('mall_customers.csv')
```

```python
mall.head()
```

```python
mall = mall.drop(columns = 'customer_id')
```

```python
mall.head()
```

```python
mall['is_male'] = pd.get_dummies(mall.gender, drop_first=True)
```

```python
mall.head()
```

```python
age_income_cluster = mall[['age', 'annual_income']]
```

```python
age_income_cluster.head()
```

```python
mall_1_inertia = []

for n in range(1,11):
    
    kmeans = KMeans(n_clusters=n)
    kmeans.fit(age_income_cluster)
    mall_1_inertia.append(kmeans.inertia_)
```

```python
plt.plot(np.arange(1,11,), mall_1_inertia)

plt.xticks(np.arange(1,11,step=1))

plt.title('Change in Inertia')

plt.show()
```

```python
kmeans = KMeans(n_clusters=4)

kmeans.fit(age_income_cluster)

kmeans.predict(age_income_cluster)
```

```python
mall['age_income_cluster']  = kmeans.predict(age_income_cluster)
```

```python
mall.head()
```

```python
sns.scatterplot(mall, x = 'age', y='annual_income', hue='age_income_cluster', palette= 'rainbow')

plt.show()
```

## 3. How does scaling impact the results of clustering?

```python
mall = mall.drop(columns = 'gender')
```

```python
mall.head()
```

```python
scaler = MinMaxScaler().set_output(transform='pandas')
```

```python
mall_scaled = scaler.fit_transform(mall)
```

```python
age_income_scale = mall_scaled[['age', 'annual_income']]
```

```python
kmeans = KMeans(n_clusters=4)

kmeans.fit(age_income_scale)

kmeans.predict(age_income_scale)
```

```python
mall['age_income_scale_cluster'] = kmeans.predict(age_income_scale)
```

```python

sns.scatterplot(mall, x = 'age', y='annual_income', hue='age_income_scale_cluster', palette= 'rainbow')

plt.show()
```
